```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.books.com.tw/web/sys_saletopb/books'
webpage = requests.get(url)
soup = BeautifulSoup(webpage.text, 'html.parser')
books = soup.find('div','mod type02_m035 clearfix')
lis = books.find_all('li','item')
for li in lis:
    author = li.find('ul',{'class':'msg'}).a.text
    book = li.find('h4').a.text
    top = li.find('strong',{'class':'no'}).text
    print('排行',top)
    print('書名:',book)
    print('作者:',author,'\n')
```
