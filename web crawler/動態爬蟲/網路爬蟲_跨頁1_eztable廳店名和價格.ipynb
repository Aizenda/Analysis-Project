```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
browser = webdriver.Chrome()
browser.get("https://tw.eztable.com/search")
time.sleep(3)
for i in range(2):
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(4)
soup = BeautifulSoup(browser.page_source,'html.parser')
content = soup.find_all('div','SearchResult__StyledRestaurantContent-alt2vy-4 ydffj')
count = 0
for i in content:
    count+=1
    restaurant = i.find('h2','SearchResult__StyledName-alt2vy-7 kjEOTT').text
    price = i.find('div','SearchResult__StyledPrice-alt2vy-13 ghERpV').text
    print(count,'.',restaurant,'價格',price)
```
